```python
#Attempt to determine Markowitz Efficient Frontier with monte carlo sim of portfolio weights
```

```python
yr_returns = returns['2018-11-01':'2019-11-01']
```

```python
#out of curiousity how normal are btc log returns among other cryptos
btc_prbplt = stats.probplot(yr_returns.btc, plot=plt)
```

```python
# calculate daily and annual returns of the stocks
returns_daily = yr_returns
returns_annual = yr_returns.mean() * 250

# get daily and covariance of returns of the stock
cov_daily = returns_daily.cov()
cov_annual = cov_daily * 250
```

```python
#List for storing the returns, volatility and weights sim portfolios
port_returns = []
port_volatility = []
sharpe_ratio = []
stock_weights = []

#set number of simulated portfolios
num_assets = len(list(yr_returns))
num_portfolios = 50000

# populate the empty lists with each portfolios returns,risk and weights
for single_portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    returns = np.dot(weights, returns_annual)
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_annual, weights)))
    sharpe = returns / volatility
    sharpe_ratio.append(sharpe)
    port_returns.append(returns)
    port_volatility.append(volatility)
    stock_weights.append(weights)
```

```python
# ceate a dict for the returns vol and sharpe
portfolio = {'Returns': port_returns,
             'Volatility': port_volatility,
             'Sharpe Ratio': sharpe_ratio}

# extend original dictionary to accomodate each ticker and weight in the portfolio
for counter,symbol in enumerate(list(yr_returns)):
    portfolio[symbol+' Weight'] = [Weight[counter] for Weight in stock_weights]

df = pd.DataFrame(portfolio)

# rename labels in the df
column_order = ['Returns', 'Volatility', 'Sharpe Ratio'] + [stock+' Weight' for stock in list(yr_returns)]

# reorder dataframe columns
df = df[column_order]



# find min Volatility & max sharpe values in the dataframe (df)
min_volatility = df['Volatility'].min()
max_sharpe = df['Sharpe Ratio'].max()

# use the min, max values to locate and create the two special portfolios
sharpe_portfolio = df.loc[df['Sharpe Ratio'] == max_sharpe]
min_variance_port = df.loc[df['Volatility'] == min_volatility]

# plot frontier, max sharpe & min Volatility values with a scatterplot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio',
                cmap='RdYlGn', edgecolors='black', figsize=(10, 8), grid=True)
plt.scatter(x=sharpe_portfolio['Volatility'], y=sharpe_portfolio['Returns'], c='red', marker='D', s=200)
plt.scatter(x=min_variance_port['Volatility'], y=min_variance_port['Returns'], c='blue', marker='D', s=200 )
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```

### Value at Risk

```python
import datetime
import numpy as np
from scipy.stats import norm


def var_cov_var(P, c, mu, sigma):
    """
    Variance-Covariance calculation of daily Value-at-Risk
    using confidence level c, with mean of returns mu
    and standard deviation of returns sigma, on a portfolio
    of value P.
    """
    alpha = norm.ppf(1-c, mu, sigma)
    return P - P*(alpha + 1)
```

```python
#Define a Function to be used in a package later on..

def run_montecarlo_pricesim(price, time_prd):
    #calculate the compound annual growth rate (CAGR) which 
    #will give us our mean return input (mu) 
    #days = (price.index[-1] - price.index[0]).days
    len(price)
    cagr = ((((price[-1]) / price[1])) ** (252.0/days)) - 1
    print ('CAGR =',str(round(cagr,4)*100)+"%")
    mu = cagr
    x_ret =  np.log(price/price.shift(1))
    vol = x_ret.std() * np.sqrt(252)
    print ("Annual Volatility =",str(round(vol,4)*100)+"%")
    
    
    #Define Variables
    S = price[-1] #starting stock price (i.e. last available real stock price)
    T = time_prd #Number of trading days
    #mu = 0.2309 #Return
    #vol = 0.4259 #Volatility

    #create list of daily returns using random normal distribution
    daily_returns=np.random.normal((mu/T),vol/np.sqrt(T),T)+1

    #set starting price and create price series generated by above random daily returns
    price_list = [S]

    for x in daily_returns:
        price_list.append(price_list[-1]*x)
    sim_ps = pd.Series(price_list)
    sim_ret =np.log(sim_ps / sim_ps.shift(1))
    print ("Simulation Log Returns =",str(round(sim_ret.sum(),4)))
    #Generate Plots - price series and histogram of daily returns
           
    plt.plot(price_list)
    plt.show()
    plt.hist(daily_returns-1, 100) #Note that we run the line plot and histogram separately, not simultaneously.
    plt.show()
```

```python

price = pd.Series(coins_mkt['close_btc'])

rets = np.log(price/price.shift(1))
rets = rets[1:]

daily_vol = rets.std()
daily_ret = rets.mean()
    
```

```python
simulation_df = pd.DataFrame()

num_simulations = 1000  #takes a while if doing lots
predicted_days = 60

last_price = price.iloc[-1]

#Create Each Simulation as a Column in df
for x in range(num_simulations):
    count = 0  
    price_series = []
    price_series.append(last_price)
        
    #Series for Preditcted Days
    for i in range(predicted_days):
        if count == 251:
            break
        
        #price = price_series[count] + price_series[count] * (daily_ret + daily_vol * np.random.normal(0, 1))
        
        price = price_series[count] * (1 + np.random.normal(0, daily_vol))

        price_series.append(price)
        count += 1

    simulation_df[x] = price_series

plt.plot(simulation_df)
plt.show()
```

```python
price_array = simulation_df.iloc[-1, :]
price_array = sorted(price_array, key=int) 

var0 = np.percentile(price_array, .90)
var =  np.percentile(price_array, .95)
var1 =  np.percentile(price_array, .99)
var2 =  np.percentile(price_array, .9999)
```

```python

print("VaR at 90% Confidence: " + '${:,.2f}'.format(last_price - var0))
print("VaR at 95% Confidence: " + '${:,.2f}'.format(last_price - var))
print("VaR at 99% Confidence: " + '${:,.2f}'.format(last_price - var1))
print("VaR at 99.99% Confidence: " + '${:,.2f}'.format(last_price - var2))
```

```python
price = pd.Series(coins_mkt['close_btc'])
price.head()
```

```python
btc_ps = pd.DataFrame(price).reset_index()
btc_ps.columns = ['ds','y']
btc_ps.head()
```

```python
import fbprophet as fb
```

```python
model = fb.Prophet(weekly_seasonality=True, yearly_seasonality=True, daily_seasonality=True)
model.add_seasonality('self_define_cycle', period=20, fourier_order=4)
```

```python
train = btc_ps[pd.to_datetime(btc_ps.ds) <= pd.to_datetime('2019-08-01')]
test = btc_ps[pd.to_datetime(btc_ps.ds) > pd.to_datetime('2019-08-01')]
```

```python
testing = test.set_index('ds')
testing.index = pd.to_datetime(testing.index)
training = train.set_index('ds')
training.index = pd.to_datetime(training.index)
predict_period = len(testing)
```

```python
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=predict_period)
```

```python
fcast = model.predict(future)
```

```python
model.plot(fcast)
plt.plot(testing.index, testing.values, '.', color='#ff3333', alpha=0.6)
plt.xlabel('Date', fontsize=12, fontweight='bold', color='gray')
plt.ylabel('Price', fontsize=12, fontweight='bold', color='gray')
plt.show()
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import median_absolute_error
```

```python
ret = max(fcast.self_define_cycle) - min(fcast.self_define_cycle)
model_df = fcast['yhat']
model_df.index = fcast['ds'].astype('datetime64[ns]')
out_df = pd.concat([testing, model_df], axis=1)
out_df = out_df[~out_df.iloc[:, 0].isnull()]
out_df = out_df[~out_df.iloc[:, 1].isnull()]
mse = mean_squared_error(out_df.iloc[:, 0], out_df.iloc[:, 1])
mae = mean_absolute_error(out_df.iloc[:, 0], out_df.iloc[:, 1])
r2 = r2_score(out_df.iloc[:, 0], out_df.iloc[:, 1])
mslog = mean_squared_log_error(out_df.iloc[:, 0], out_df.iloc[:, 1])
mae = median_absolute_error(out_df.iloc[:, 0], out_df.iloc[:, 1])

rep = [ret, mse, mae, r2, mslog, mae]
print("Projected return per cycle: {}".format(round(rep[0], 2)))
print("MSE:  {}".format(round(rep[1], 4)))
print("MAE:  {}".format(round(rep[2], 4)))
print("R2:   {}".format(round(rep[3], 4)))
print("MSLE: {}".format(round(rep[4], 4)))
print("MAE:  {}".format(round(rep[5], 4)))
```
